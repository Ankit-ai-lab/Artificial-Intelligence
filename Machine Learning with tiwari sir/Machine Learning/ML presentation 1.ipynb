# BI-VARIANT.
- CATEGORY AND REVIEW.

```python
sns.barplot(x='Category', y='Review Rating', data=shop, errorbar=None)
plt.title('Bivariate Bar Plot')
plt.xlabel('Category')
plt.ylabel('Review Rating')
plt.show()
```

- LOCATION AND FREQUENCY OF PURCHASE

```python
sns.stripplot(data=shop, x='Gender', y='Frequency of Purchases')
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='Discount Applied',y='Purchase Amount (USD)',data=shop)
plt.title('Box plot with Seaborn')
plt.xlabel("Discount Applied")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```
