```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values_1 = [25, 40, 30, 50]  # First variable
values_2 = [15, 30, 25, 35]  # Second variable
values_3 = [10, 20, 15, 25]  # Third variable

# Creating a DataFrame for Seaborn
import pandas as pd
df = pd.DataFrame({
    'Category': categories * 3,
    'Variable': ['Variable 1'] * len(categories) + ['Variable 2'] * len(categories) + ['Variable 3'] * len(categories),
    'Value': values_1 + values_2 + values_3
})

# Creating a grouped bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', hue='Variable', data=df, palette='Set3')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Trivariant Column Bar Chart with Seaborn')

# Displaying the chart
plt.show()
```
